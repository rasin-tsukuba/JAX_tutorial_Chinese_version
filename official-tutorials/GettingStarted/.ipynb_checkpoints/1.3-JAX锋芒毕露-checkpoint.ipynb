# ğŸ”ªçº¯å‡½æ•°

JAXå˜æ¢å’Œç¼–è¯‘ä»…é€‚ç”¨äºåŠŸèƒ½çº¯å‡€çš„Pythonå‡½æ•°ï¼šæ‰€æœ‰è¾“å…¥æ•°æ®å‡é€šè¿‡å‡½æ•°å‚æ•°ä¼ é€’ï¼Œæ‰€æœ‰ç»“æœå‡é€šè¿‡å‡½æ•°ç»“æœè¾“å‡ºã€‚å¦‚æœä½¿ç”¨ç›¸åŒçš„è¾“å…¥è°ƒç”¨çº¯å‡½æ•°ï¼Œåˆ™æ€»ä¼šè¿”å›ç›¸åŒçš„ç»“æœã€‚

è¿™æ˜¯ä¸€äº›åŠŸèƒ½ä¸Šå¹¶éçº¯å‡½æ•°çš„ç¤ºä¾‹ï¼Œå¯¹äºè¿™äº›å‡½æ•°ï¼ŒJAXçš„è¡Œä¸ºä¸åŒäºPythonè§£é‡Šå™¨ã€‚æ³¨æ„ï¼ŒJAXç³»ç»Ÿä¸èƒ½ä¿è¯è¿™äº›è¡Œä¸ºã€‚ä½¿ç”¨JAXçš„æ­£ç¡®æ–¹æ³•æ˜¯ä½¿ç”¨åŠŸèƒ½çº¯ç²¹çš„çº¯Pythonå‡½æ•°ã€‚

```python
def impure_print_side_effect(x):
    print("Executing function") # è¿™å°±æ˜¯ä¸€ç§å‰¯ä½œç”¨
    return x

# The side-effects appear during the first run  
# åœ¨ç¬¬ä¸€æ¬¡æ‰§è¡Œæ—¶å°±å‡ºç°å‰¯ä½œç”¨
print("First call: ", jit(impure_print_side_effect)(4.))

# Subsequent runs with parameters of same type and shape may not show the side-effect
# This is because JAX now invokes a cached compilation of the function
# ä½¿ç”¨ç›¸åŒç±»å‹å’Œå½¢çŠ¶çš„å‚æ•°è¿›è¡Œçš„åç»­è¿è¡Œå¯èƒ½ä¸ä¼šæ˜¾ç¤ºå‰¯ä½œç”¨
# è¿™æ˜¯å› ä¸ºJAXç°åœ¨è°ƒç”¨äº†è¯¥å‡½æ•°çš„ç¼“å­˜ç¼–è¯‘
print("Second call: ", jit(impure_print_side_effect)(5.))

# JAX re-runs the Python function when the type or shape of the argument changes
# å½“å‚æ•°çš„ç±»å‹æˆ–è€…å½¢çŠ¶æ›´æ”¹æ—¶ï¼ŒJAXé‡æ–°è¿è¡ŒPythonå‡½æ•°
print("Third call, different type: ", jit(impure_print_side_effect)(jnp.array([5.])))
```

```python
g = 0.
def impure_uses_globals(x):
  return x + g

# JAX captures the value of the global during the first run
# JAXåœ¨ç¬¬ä¸€æ¬¡æ‰§è¡Œæ—¶æ•æ‰åˆ°å…¨å±€å˜é‡
print ("First call: ", jit(impure_uses_globals)(4.))

g = 10.  # æ›´æ–°å…¨å±€å˜é‡

# Subsequent runs may silently use the cached value of the globals
# ä»¥ä¸‹çš„ç»“æœå°†ä¼šé»˜è®¤ä½¿ç”¨ç¼“å­˜çš„å…¨å±€å˜é‡
print ("Second call: ", jit(impure_uses_globals)(5.))

# JAX re-runs the Python function when the type or shape of the argument changes
# This will end up reading the latest value of the global
# å‚æ•°çš„ç±»å‹æˆ–è€…å½¢çŠ¶æ›´æ”¹æ—¶ï¼ŒJAXé‡æ–°è¿è¡ŒPythonå‡½æ•°
# è¿™æ ·å°†ä¼šé‡æ–°è¯»å–æœ€æ–°çš„å…¨å±€å˜é‡å€¼
print ("Third call, different type: ", jit(impure_uses_globals)(jnp.array([4.])))
```

```python
g = 0.
def impure_saves_global(x):
  global g
  g = x
  return x

# JAX runs once the transformed function with special Traced values for arguments
# JAXè¿è¡Œå¸¦æœ‰å‚æ•°çš„ç‰¹æ®Šè·Ÿè¸ªå€¼è½¬æ¢åçš„å‡½æ•°
print ("First call: ", jit(impure_saves_global)(4.))
print ("Saved global: ", g)  # ä¿å­˜çš„å…¨å±€å˜é‡å¸¦æœ‰å†…éƒ¨çš„JAXå€¼
```

å³æ—¶Pythonå‡½æ•°å®é™…ä¸Šåœ¨å†…éƒ¨ä½¿ç”¨æœ‰çŠ¶æ€çš„å¯¹è±¡ï¼Œåªè¦å®ƒä¸è¯»å–æˆ–å†™å…¥å¤–éƒ¨çŠ¶æ€ï¼Œå®ƒåœ¨åŠŸèƒ½ä¸Šä¹Ÿå¯ä»¥æ˜¯çº¯å‡½æ•°ï¼š

```python
def pure_uses_internal_state(x):
    state = dict(even=0, odd=0)
    for i in range(10):
        state['even' if i % 2 == 0 else 'odd'] += x
    return state['even'] + state['odd']

print(jit(pure_uses_internal_state)(5.))
```

ä¸å»ºè®®åœ¨éœ€è¦ `jit` çš„ä»»ä½•JAXå‡½æ•°æˆ–ä»»ä½•æ§åˆ¶æµåŸè¯­ä¸­ä½¿ç”¨è¿­ä»£å™¨ã€‚å› ä¸ºè¿­ä»£å™¨æ˜¯ä¸€ä¸ªPythonå¯¹è±¡ï¼Œå®ƒå¼•å…¥çŠ¶æ€ä»¥æ£€æµ‹ä¸‹ä¸€ä¸ªå…ƒç´ ã€‚å› æ­¤ï¼Œå®ƒä¸JAXåŠŸèƒ½ç¼–ç¨‹æ¨¡å‹ä¸å…¼å®¹ã€‚ä¸‹é¢çš„ä»£ç ä¸­æœ‰ä¸€äº›å°†è¿­ä»£å™¨ä¸JAXä¸€èµ·ä½¿ç”¨çš„é”™è¯¯ç¤ºä¾‹ã€‚å®ƒä»¬ä¸­çš„å¤§å¤šæ•°è¿”å›é”™è¯¯ï¼Œä½†æœ‰äº›ä¼šäº§ç”Ÿä»¤äººæ„å¤–çš„ç»“æœï¼š

```python
import jax.numpy as jnp
import jax.lax as lax
from jax import make_jaxpr

# lax.fori_loop
array = jnp.arange(10)
print(lax.fori_loop(0, 10, lambda i, x: x+array[i], 0)) # é¢„æœŸç»“æœæ˜¯45
iterator = iter(range(10))
print(lax.fori_loop(0, 10, lambda i, x:x+next(iterator), 0)) #æ„å¤–ç»“æœæ˜¯0

#lax.scan
def func11(arr, extra):
    ones = jnp.ones(arr.shape)
    def body(carry, aelems):
        ae1, ae2 = aelems
        return (carry + ae1 * ae2 + extra, carry)
    return lax.scan(body, 0., (arr, ones))
make_jaxpr(func11)(jnp.arange(16), 5.)

#lax.cond
array_operand = jnp.array([0.])
lax.cond(True, lambda x: x+1, lambda x:x-1, array_operand)
iter_operand = iter(range(10))
# æŠ¥é”™
lax.cond(True, lambda x: next(x)+1, lambda x:next(x)-1, iter_operand)
```

## ğŸ”ª å°±åœ°æ›´æ–°

åœ¨Numpyä¸­æˆ‘ä»¬ä¹ æƒ¯è¿™ä¹ˆåšï¼š

```python
numpy_array = np.zeros((3, 3), dtype=np.float32)
print("original array:")
print(numpy_array)

# In place, mutating update
# ç›´æ¥å¯å˜æ›´æ–°
numpy_array[1, :] = 1.0
print("updated array:")
print(numpy_array)
```

ä½†æ˜¯ï¼Œå¦‚æœå°è¯•å°±åœ°æ›´æ–°JAXæ•°ç»„ï¼Œåˆ™ä¼šæ”¶åˆ°é”™è¯¯æ¶ˆæ¯ï¼ ï¼ˆâ˜‰_â˜‰ï¼‰

```python
jax_array = jnp.zeros((3,3), dtype=jnp.float32)

# In place update of JAX's array will yield an error!
# å°±åœ°æ›´æ–°JAXæ•°ç»„åˆ™ä¼šå¯¼è‡´é”™è¯¯
try:
  jax_array[1, :] = 1.0
except Exception as e:
  print("Exception {}".format(e))
```

JAXæä¾›äº†å‡ ä¸ªå‡½æ•°å¼çš„æ›´æ–°å‡½æ•°ï¼š[index_update](https://jax.readthedocs.io/en/latest/_autosummary/jax.ops.index_update.html#jax.ops.index_update), [index_add](https://jax.readthedocs.io/en/latest/_autosummary/jax.ops.index_add.html#jax.ops.index_add), [index_min](https://jax.readthedocs.io/en/latest/_autosummary/jax.ops.index_max.html#jax.ops.index_min), [index_max](https://jax.readthedocs.io/en/latest/_autosummary/jax.ops.index_max.html#jax.ops.index_max), ä»¥åŠ [index](https://jax.readthedocs.io/en/latest/_autosummary/jax.ops.index.html#jax.ops.index)è¾…åŠ©å‡½æ•°ã€‚

ï¸âš ï¸ åœ¨`jit`ä»£ç çš„ `lax.while_loop`æˆ– `lax.fori_loop`ä¸­ï¼Œåˆ‡ç‰‡çš„å¤§å°ä¸èƒ½åšä¸ºå‚æ•°å€¼çš„å‡½æ•°ï¼Œè€Œåªèƒ½æ˜¯å‚æ•°å½¢çŠ¶çš„å‡½æ•°â€”â€”åˆ‡ç‰‡å¼€å§‹ç´¢å¼•æ²¡æœ‰è¿™ç§é™åˆ¶ã€‚æœ‰å…³æ­¤é™åˆ¶çš„æ›´å¤šä¿¡æ¯ï¼Œè¯·å‚è§ä¸‹é¢çš„æ§åˆ¶æµéƒ¨åˆ†ã€‚

```python
from jax.ops import index, index_add, index_update
```

### index_update

å¦‚æœ **index_update**çš„**input values** æ²¡æœ‰è¢«é‡ç”¨ï¼Œ`jit`ç¼–è¯‘çš„ä»£ç å°†ä¼šå°±åœ°æ‰§è¡Œè¿™äº›æ“ä½œï¼š

```python
jax_array = jnp.zeros((3, 3))
print("original array:")
print(jax_array)

new_jax_array = index_update(jax_array, index[1, :], 1.)

print("old array unchanged:")
print(jax_array)

print("new array:")
print(new_jax_array)
```

### index_add

å¦‚æœ **index_add**çš„**input values** æ²¡æœ‰è¢«é‡ç”¨ï¼Œ`jit`ç¼–è¯‘çš„ä»£ç å°†ä¼šå°±åœ°æ‰§è¡Œè¿™äº›æ“ä½œï¼š


```python
print("original array:")
jax_array = jnp.ones((5, 6))
print(jax_array)

new_jax_array = index_add(jax_array, index[::2, 3:], 7)
print("new array post-addition:")
print(new_jax_array)
```
