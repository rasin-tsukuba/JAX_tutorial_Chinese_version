## å¦‚ä½•ä½¿ç”¨JAXå˜æ¢

åœ¨ä¸Šä¸€èŠ‚ä¸­ï¼Œæˆ‘ä»¬è®¨è®ºäº†JAXå…è®¸æˆ‘ä»¬å˜æ¢Pythonå‡½æ•°ã€‚è¿™æ˜¯é€šè¿‡é¦–å…ˆå°†Pythonå‡½æ•°è½¬æ¢ä¸ºä¸€ç§ç®€å•çš„ä¸­é—´è¯­è¨€jaxpræ¥å®Œæˆçš„ã€‚ä¹‹åï¼Œè½¬æ¢å°†åœ¨jaxprå½¢å¼ä¸Šè¿›è¡Œã€‚

æˆ‘ä»¬å¯ä»¥ç”¨ `jax.make_jaxpr` æ¥æ˜¾ç¤ºå‡½æ•°çš„jaxprå½¢å¼ï¼š

```python
import jax
import jax.numpy as jnp

global_list = []

def log2(x):
    global_list.append(x)
    ln_x = jnp.log(x)
    ln_2 = jnp.log(2.0)
    
    return ln_x / ln_2

print(jax.make_jaxpr(log2)(3.0))
```

åœ¨æ•™ç¨‹ä¸­çš„[ç†è§£Jaxprs]()éƒ¨åˆ†æä¾›äº†æœ‰å…³ä¸Šè¿°è¾“å‡ºå«ä¹‰çš„æ›´å¤šä¿¡æ¯ã€‚

è¯·æ³¨æ„ï¼Œå¾ˆé‡è¦çš„ä¸€ç‚¹æ˜¯jaxpræ— æ³•æ•è·è¯¥å‡½æ•°çš„å‰¯ä½œç”¨ï¼šå…¶ä¸­æ²¡æœ‰ä¸`global_list.append(x)`çš„å†…å®¹ã€‚è¿™æ˜¯ä¸€ä¸ªç‰¹æ€§ï¼Œå¹¶ä¸æ˜¯ä¸€ä¸ªæ¼æ´ï¼šJAXæ—¨åœ¨ç†è§£æ— å‰¯ä½œç”¨çš„ä»£ç ã€‚å¦‚æœæ‚¨ä¸å¤ªç†Ÿæ‚‰çº¯å‡½æ•°å’Œå‰¯ä½œç”¨è¿™ä¸¤ä¸ªæœ¯è¯­ï¼Œè¯·å‚è§[JAXé”‹èŠ’æ¯•éœ²:ğŸ”ªçº¯å‡½æ•°](https://render.githubusercontent.com/view/ipynb?color_mode=light&commit=fe4a5f85bf7936468ed39f20cced5b25a1612efb&enc_url=68747470733a2f2f7261772e67697468756275736572636f6e74656e742e636f6d2f726173696e2d7473756b7562612f4a41585f6368696e6573655f7475746f7269616c2f666534613566383562663739333634363865643339663230636365643562323561313631326566622f6f6666696369616c2d7475746f7269616c732f47657474696e67537461727465642f312e332d4a41582545392539342538422545382538412539322545362541462539352545392539432542322e6970796e62&nwo=rasin-tsukuba%2FJAX_chinese_tutorial&path=official-tutorials%2FGettingStarted%2F1.3-JAX%E9%94%8B%E8%8A%92%E6%AF%95%E9%9C%B2.ipynb&repository_id=349726397&repository_type=Repository#%F0%9F%94%AA%E7%BA%AF%E5%87%BD%E6%95%B0)ã€‚

å½“ç„¶ï¼Œéçº¯å‡½æ•°ä»ç„¶å¯ä»¥ç¼–å†™ç”šè‡³è¿è¡Œï¼Œä½†æ˜¯ä¸€æ—¦è½¬æ¢ä¸ºjaxprï¼ŒJAXå°±æ— æ³•ä¿è¯å…¶è¡Œä¸ºã€‚ä½†æ ¹æ®ç»éªŒï¼Œæ‚¨å¯ä»¥æœŸæœ›ï¼ˆä½†ä¸åº”è¯¥ä¾èµ–ï¼‰JAXè½¬æ¢å‡½æ•°çš„å‰¯ä½œç”¨åªè¿è¡Œä¸€æ¬¡ï¼ˆåœ¨ç¬¬ä¸€æ¬¡è°ƒç”¨æ—¶ï¼‰ä¹‹åå†ä¹Ÿä¸ä¼šè¿è¡Œã€‚è¿™æ˜¯å› ä¸ºJAXä½¿ç”¨ç§°ä¸ºâ€œè·Ÿè¸ªâ€çš„è¿‡ç¨‹ç”Ÿæˆjaxprçš„æ–¹å¼ã€‚

è·Ÿè¸ªæ—¶ï¼ŒJAXç”¨è·Ÿè¸ªå™¨å¯¹è±¡åŒ…è£…æ¯ä¸ªå‚æ•°ã€‚ç„¶åï¼Œè¿™äº›è·Ÿè¸ªå™¨è®°å½•å‡½æ•°è°ƒç”¨æœŸé—´å¯¹ä»–ä»¬æ‰§è¡Œçš„æ‰€æœ‰JAXæ“ä½œï¼ˆå‘ç”Ÿåœ¨Pythonä»£ç ä¹‹ä¸­ï¼‰ã€‚ä¹‹åï¼ŒJAXä½¿ç”¨è·Ÿè¸ªè®°å½•æ¥é‡æ„æ•´ä¸ªå‡½æ•°ã€‚è¯¥é‡å»ºçš„è¾“å‡ºæ˜¯jaxprã€‚ç”±äºè·Ÿè¸ªå…¶æ²¡æœ‰è®°å½•Pythonçš„å‰¯ä½œç”¨ï¼Œå› æ­¤å®ƒä»¬ä¸ä¼šå‡ºç°åœ¨jaxprä¸­ã€‚ä½†æ˜¯ï¼Œå‰¯ä½œç”¨ä»ä¼šåœ¨è·Ÿè¸ªæœŸé—´å‘ç”Ÿã€‚

æ³¨æ„ï¼šPythonçš„ `print()` ä¸æ˜¯çº¯å‡½æ•°ï¼šæ–‡æœ¬è¾“å‡ºæ˜¯è¯¥å‡½æ•°çš„å‰¯ä½œç”¨ã€‚å› æ­¤ï¼Œä»»ä½• `print()`è°ƒç”¨éƒ½åªä¼šåœ¨è·Ÿè¸ªè¿‡ç¨‹ä¸­å‘ç”Ÿï¼Œè€Œä¸ä¼šå‡ºç°åœ¨jaxprä¸­ï¼š

```python
def log2_with_print(x):
    print("printed x: ", x)
    ln_x = jnp.log(x)
    ln_2 = jnp.log(2)
    return ln_x / ln_2

print(jax.make_jaxpr(log2_with_print)(3.))
```

çœ‹åˆ°æ‰“å°çš„`x`æˆä¸ºä¸€ä¸ª `Traced` å¯¹è±¡äº†å—ï¼Ÿè¿™å°±æ˜¯JAXçš„å†…éƒ¨è¿è¡Œæœºåˆ¶ã€‚

Pythonä»£ç è‡³å°‘è¿è¡Œä¸€æ¬¡çš„äº‹å®ä¸¥æ ¼ä¸Šæ¥è¯´æ˜¯å®ç°ç»†èŠ‚ï¼Œå› æ­¤ä¸åº”è¯¥å¯¹å…¶æœ‰ä¾èµ–ã€‚ä½†æ˜¯ï¼Œç†è§£å®ƒå¾ˆæœ‰ç”¨ï¼Œå› ä¸ºæ‚¨å¯ä»¥è°ƒè¯•ä»¥æ‰“å°å‡ºè®¡ç®—çš„ä¸­é—´å€¼æ—¶ä½¿ç”¨å®ƒã€‚

å…³é”®è¦ç†è§£çš„æ˜¯ï¼Œjaxprä¼šæ•è·å¯¹ç»™å®šå‚æ•°æ‰§è¡Œçš„åŠŸèƒ½ã€‚ä¾‹å¦‚ï¼Œå¦‚æœæˆ‘ä»¬æœ‰æ¡ä»¶ï¼Œé‚£ä¹ˆjaxprå°†åªçŸ¥é“æˆ‘ä»¬é‡‡å–çš„åˆ†æ”¯ï¼š

```python
def log2_if_rank_2(x):
    if x.ndim == 2:
        ln_x = jnp.log(x)
        ln_2 = jnp.log(2)
        return ln_x / ln_2
    else:
        return x
    
print(jax.make_jaxpr(log2_if_rank_2)(jax.numpy.array([1., 2., 3.])))
```

## ä½¿ç”¨JITç¼–è¯‘å‡½æ•°

å¦‚å‰æ‰€è¿°ï¼ŒJAXä½¿æ“ä½œå¯ä»¥ä½¿ç”¨ç›¸åŒçš„çš„ä»£ç åœ¨CPUã€GPUå’ŒTPUä¸Šæ‰§è¡Œã€‚è®©æˆ‘ä»¬æ¥çœ‹ä¸€ä¸ªè®¡ç®—æ¯”ä¾‹æŒ‡æ•°çº¿æ€§å•ä½ï¼ˆSELUï¼‰çš„ç¤ºä¾‹ï¼Œè¿™æ˜¯æ·±åº¦å­¦ä¹ ä¸­å¸¸ç”¨çš„ä¸€ç§è¿ç®—ï¼š

```python
import jax
import jax.numpy as jnp

def selu(x, alpha=1.67, lambda_=1.05):
    return lambda_ * jnp.where(x > 0, x, alpha * jnp.exp(x) - alpha)

x = jnp.arange(1000000)
%timeit selu(x).block_until_ready()
```

ä»¥ä¸Šä»£ç ä¸€æ¬¡æ€§å‘åŠ é€Ÿå™¨å‘é€äº†ä¸€ä¸ªæ“ä½œã€‚è¿™é™åˆ¶äº†XLAç¼–è¯‘å™¨ä¼˜åŒ–åŠŸèƒ½çš„èƒ½åŠ›ã€‚

è‡ªç„¶ï¼Œæˆ‘ä»¬æƒ³è¦åšçš„äº‹ç»™XLAç¼–è¯‘å™¨å°½å¯èƒ½å¤šçš„ä»£ç ï¼Œä»¥ä¾¿å®ƒå¯ä»¥å®Œå…¨ä¼˜åŒ–å®ƒã€‚ä¸ºæ­¤ï¼ŒJAXæä¾›äº† `jax.jit`è½¬æ¢ï¼Œå®ƒå°†å³æ—¶ç¼–è¯‘JAXå…¼å®¹çš„å‡½æ•°ã€‚ä¸‹é¢çš„ç¤ºä¾‹æ˜¾ç¤ºäº†å¦‚ä½•ä½¿ç”¨JITæ¥åŠ å¿«æ­¤å‡½æ•°ï¼š

```python
selu_jit = jax.jit(selu)

# warm up
#é¢„çƒ­
selu_jit(x).block_until_ready()

%timeit selu_jit(x).block_until_ready()
```

ä»¥ä¸‹æ˜¯åˆšæ‰å‘ç”Ÿäº‹æƒ…çš„è¯¦ç»†è§£é‡Šï¼š

1. æˆ‘ä»¬å°†`selu_jit`å®šä¹‰ä¸º`selu`çš„ç¼–è¯‘ç‰ˆæœ¬
2. æˆ‘ä»¬åœ¨`x`ä¸Šè¿è¡Œä¸€æ¬¡ `selu_jit`ã€‚è¿™å°±æ˜¯JAXè¿›è¡Œè·Ÿè¸ªçš„åœ°æ–¹â€”â€”æ¯•ç«Ÿä»–éœ€è¦ä¸€äº›è¾“å…¥æ‰èƒ½åŒ…è£…åœ¨è·Ÿè¸ªå™¨ä¸­ã€‚ç„¶åï¼Œä½¿ç”¨XLAå°†jaxprç¼–è¯‘ä¸ºé’ˆå¯¹æ‚¨çš„GPUæˆ–TPUä¼˜åŒ–çš„éå¸¸æœ‰æ•ˆçš„ä»£ç ã€‚ç°åœ¨ï¼Œå¯¹`selu_jit`çš„åç»­è°ƒç”¨å°†ä½¿ç”¨æ”¹ä»£ç ï¼Œä»è€Œå®Œå…¨è·³è¿‡æˆ‘ä»¬ä»¥å‰çš„Pythonå®ç°ã€‚
